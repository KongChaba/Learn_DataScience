# 1.Business Understanding
วิเคราะห์ข้อมูลคลิปวิดีโอ Youtube ช่อง [ออกกำลังกาย](https://www.youtube.com/channel/UCVQJZE_on7It_pEv6tn-jdA) ซึ่งเป็นช่องสำหรับคนที่รักสุขภาพและต้องการออกกำลังกายอย่างสม่ำเสมอ <br> ในช่องจะโพสคลิปวีดีโอทุกวัน  มีผู้ติดตามมากกว่า 7 แสน subscribers ยอดผู้ติดตามเพิ่มขึ้นอย่างรวดเร็ว <br>จึงต้องการวิเคราะห์ข้อมูลเพื่อ ต้องการทราบ ลักษณะของคลิปวีดีโอที่โพส, พฤติกรรมของผู้ติดตามคลิป

*   คนมักออกกำลังกายช่วงเย็นวันศุกร์ และเสาร์ อาทิตย์ ทำให้มียอดวิวสูงขึ้น ดังนั้นถ้าอยากให้มีผู้ติดตามอย่างสม่ำเสมอ ควรโพสวีดีโอ ช่วงเช้าวันศุกร์
*   ผู้ติดตามคลิปน่าจะเป็นคนที่ออกกำลังกายด้วยตนเองที่บ้าน น่าจะชอบออกกำลังกายตามคลิปที่มีความยาวไม่เกิน 30 นาที 
*   คลิปที่มียอดวิวเยอะที่สุด คือคลิปแนะนำตัว

```
df3 = df.copy()
```

```
df3.head(3)
```

```
df3['length'].describe()
```

```
sns.displot(data=df3, x='length', bins=5)
plt.show()
```

```
#convert numerical to ordinal
label_length = ['Short', 'Medium', 'Long', 'Very long']
cut_bins = [0, 20 ,40, 50, 100]

df3['length_level'] = pd.cut(df3['length'], bins=cut_bins, labels=label_length)
```

```
df3.sample(3)
```

```
avg_short = round(df3[df3['length_level']=='Short']['calories'].mean(), 0)
avg_short = 0
avg_medium = round(df3[df3['length_level']=='Medium']['calories'].mean(), 0)
avg_long = round(df3[df3['length_level']=='Long']['calories'].mean(), 0)
avg_very_long = round(df3[df3['length_level']=='Very long']['calories'].mean(), 0)
```

```
print(avg_short)
print(avg_medium)
print(avg_long)
print(avg_very_long)
```

```
df3[df3['calories'].isnull()].head(3)
```

```
idx = df3[df3['calories'].isnull()].index
```

```
df3.loc[[1]]
```

```
#ใช้สำหรับเข้าถึง แถวนั้นเลย
df3.loc[[1,9,10], ['length_level']]
```

```
#ใช่สำหรับเข้าถึงเซลล์นั้นๆเลย
df3.iat[1,3]
```

```
df3.iat[1,9]
```

```
for x in idx:
  #print(df3.loc[x, 'length_level'])
  if df3.loc[x, 'length_level'] == 'Short':
    df3.iat[x, 3] = avg_short
  elif df3.loc[x, 'length_level'] == 'Medium':
    df3.iat[x, 3] = avg_medium
  elif df3.loc[x, 'length_level'] == 'Long':
    df3.iat[x, 3] = avg_long
  else :
    df3.iat[x,3] = avg_very_long
```

## ข้อ 14 ยอดคนดูคลิปในแต่ละวัน แตกต่างกันอย่างไร

```
df3.sample(3)
```

```
df3['day'] = df3['dayofweek'].replace(to_replace=replace_values)
```

```
df3.head(3)
```

```
plt.figure(figsize=(12,7))
sns.boxplot(data=df3, x='day', y='views')
```

```
plt.figure(figsize=(12,7))
sns.boxplot(data=df3[df3['views']<200000],
            x='day',
            y='views',
            order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
```
