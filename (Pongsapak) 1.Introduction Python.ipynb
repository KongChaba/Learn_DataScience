```python
#slide63 celsius to fahrenheit & kevin
def temp(C):
  F = (9/5)*C + 32
  K = C + 273.15
  return F, K

C = float(input("Input celsius : "))
F, H = temp(C)
print("{:.2f} Celsius = {:.2f} Fahrenheit = {:.2f} Kevin".format(C, F, H))
```

```python
def grade(point):
  if point >= 80 & point <= 100:
    return 'A'
  elif point >= 70 & point <= 79:
    return 'B'
  elif point >= 60 & point <= 69:
    return 'C'
  elif point >= 50 & point <= 59:
    return 'D'
  elif point >= 0 & point <= 49:
    return 'F'
  else:
    return 'Error'
    
T1 = int(input("กลางภาค : "))
T2 = int(input("ปลายภาค : "))
point = T1 + T2

p = grade(point)
print("Your grade is : {}".format(p))
```

```python
#slide84
def chart(BMI):
  if BMI >= 40:
    return 'Morbidly Obese'
  elif BMI >= 30:
    return 'Obesity'
  elif BMI >= 25:
    return 'Over Weight'
  elif BMI >= 18.5:
    return 'Normal Weigh'
  elif BMI < 18.5:
    return 'Underweight'

wei_kg = input("Enter your weight (kg) : ")
hig_cm = input("Enter your height (cm) : ")
hig_m = float(hig_cm)/100
BMI =float(wei_kg) / (hig_m**2)

a = chart(int(BMI))
print("-"*30)
print("BMI : {:.1f}".format(BMI))
print(a)
```

```python
def for_string():
  school = "Hello"
  i = 1
  for c in school:
    print(f"{i}.\t {c}")
    i+=1

for_string()
```

```python
#100 Multiplication table
def time(num):
  print('-'*30)
  for i in range(1,13):
    print(f"{num} x {i} = {i*num}")
  print('-'*30)

num = int(input("Enter number : "))
time(num)
```

```python
#101
def mult_table(n,m):
  for step in range(n,m+1):
    print("-"*30)
    for i in range(1,13):
      print(f"{step} x {i} = {step*i}")
  print("-"*30)

num1 = int(input("start number : "))
num2 = int(input("end number : "))
mult_table(num1, num2)
```

```python
#103
def num_table(num):
  for i in range(1, num+1):
    print(f"{i}\t",end=" ")
    if i % 5 == 0:
      print("\n")

num = int(input("Input a number : "))
num_table(num)
```
