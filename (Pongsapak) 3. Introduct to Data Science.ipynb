### __Data Exploratory (Pivot Table)__

```python
df.head()
```

```python
df.groupby('Product')[['Price']].agg('sum').sort_values('Price',ascending=False)
#groupby('columns') ใช่้ในรวมแต่ละแถวในคอลัมน์เดียวกันที่ตรงกับคำที่ตั้งไว้
#[['Price']] ที่ต่อท้าย คือ นำมาเฉพาะคอลัมน์ Price 
#.agg('sum') โดย agg ใช้หาค่าทางคณิตศาสตร์ ซึ่งระบุ sum เพื่อหาค่าผลรวม
#.sort_values คือ เรียงค่าจากน้อยไปมาก
#ascending=False คือการการสลับการเรียงจากน้อยไปมากให้เป็นจากมากไปน้อย
```

```python
df.groupby('Product')[['Price']].agg('sum').sort_values('Price',ascending=False)
#groupby('columns') ใช่้ในรวมแต่ละแถวในคอลัมน์เดียวกันที่ตรงกับคำที่ตั้งไว้
#[['Price']] ที่ต่อท้าย คือ นำมาเฉพาะคอลัมน์ Price 
#.agg('sum') โดย agg ใช้หาค่าทางคณิตศาสตร์ ซึ่งระบุ sum เพื่อหาค่าผลรวม
#.sort_values คือ เรียงค่าจากน้อยไปมาก
#ascending=False คือการการสลับการเรียงจากน้อยไปมากให้เป็นจากมากไปน้อย
```
