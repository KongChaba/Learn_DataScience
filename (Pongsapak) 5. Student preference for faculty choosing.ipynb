# **Business Understanding**

พิจารณาข้อมูลของ นศ.คณะ IT สถาบันเทคโนโลยีไทย-ญี่ปุ่น (TNI) <br>
จำนวน 4 หลักสูตร
- IT (Information Technology)
- BI (Digital Business Information Technology)
- MT (Multimedia Technology)
- DC (Digital Communication)

ตั้งสมมตฐาน
1. นักเรียนที่มาเรียนที่ TNI มาจากโรงเรียนในกทม. เป็นสัดส่วนมากกว่า 80%
2. นักเรียน สาขา IT มีจำนวนมากที่สุด
3. ความชอบของนักศึกษาแต่ละหลักสูตร แตกต่างกัน 
    - นักศึกษา IT ชอบเขียนโค้ด มากกว่า ชอบการนำเสนอ
    - นักศึกษา BI ชอบนำเสนอ มากกว่า ชอบโค้ด-
    - นักศึกษา MT ชอบ Anime มากกว่าสาขาอื่น

```python
import math
```

```python
categories = ['Coding','Math','Apply Program','Marketing','Negotiate']
N = len(categories)

angles = [n / float(N) * 2 * math.pi for n in range(N)]
angles += angles[:1]

values=df_pivot.iloc[0].tolist()
values += values[:1]
 
# Initialise the spider plot
ax = plt.subplot(111, polar=True)

# Plot data
ax.plot(angles, values, linewidth=1, linestyle='solid')
 
# Fill area
ax.fill(angles, values, 'b', alpha=0.7, facecolor='red')

# Draw xlabels
plt.xticks(angles[:-1], categories, color='black', size=15) # Draw one axe per variable + add labels
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([1,2,3,4], ['1','2','3','4'], color='black', size=15)
plt.ylim(0,5)

plt.title('Perference of BI Department\n')
plt.show()
```
