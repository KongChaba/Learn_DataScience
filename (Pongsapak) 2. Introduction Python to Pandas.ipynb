Grading
  - 0 - 50 F
  - 51 - 60 D
  - 61 - 70 C
  - 71 - 80 B
  - 81 - 100 A

```python
def grading(x):
  if x > 80:
    return 'A'
  elif x > 70:
    return 'B'
  elif x > 60:
    return 'C'
  elif x > 50:
    return 'D'
  else :
    return 'F'

print(grading(45))
```

```python
df['score_ds'].map(grading)
```

```python
df['grade_ds'] = df['score_ds'].map(grading)
df
```

### __2.1 การเรียกดูข้อมูล (Filter data)__

#### __1. Logical Operators__

```python
#df[เงื่อนไขของ Filter]
df[df['name'] == 'Jane']
#df['name'] == 'Jane' หมายถึง เลือกแถวที่ใน column name ในแถวนั้นมีคำว่า Jane
```

```python
df[df['name'] != 'Jane']
```

```python
df[df['name'] > 'Jane']
#โปรแกรมจะเรียงลำดับชื่อตามตัวอักษรแล้วเรียกชื่อที่อยู่ถัดจากคนชื่อ Jane
```

#### __2. Multiple Logical Operators__

```python
#filter คนที่มีคะแนนมากกว่า 50 และมีอายุมากกว่า 18
df[(df['score'] > 50) & (df['age'] > 18)]
```

```python
df[(df.grade == 'B') & (df.age > 15)]
```

#### __5. Query__

```python
df.query('age > 15 and grade == "B" ')
```

#### __3. Isin__

```python
listA = ['A', 'B']
```

```python
df[ df.grade.isin(listA) ]
```

#### __4. Str accessor__

```python
df[df['name'].str.startswith('J')]
```

```python
df[df['name'].str.startswith('J')][['name', 'grade']]
```
